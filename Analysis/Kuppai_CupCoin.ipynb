```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load the Training Dataset 

```python
train = pd.read_csv('BRA.csv') # Training set is already available
train.head()
```

### Get the Training Dataset Information

```python
t=train.info()
```

### Get the Training Dataset Stats

```python
d=train.describe()
d
```

### Graph for Unqiue Features in Training Dataset

```python
dT=d.T
dT.plot.bar(y='unique')
plt.title("Bar plot of the unique of numeric features",fontsize=17)
```

### Size Details Stats

```python
sns.set_style('whitegrid')
sns.countplot(x='Size',data=train,palette='RdBu_r')
sns.set(rc={'figure.figsize':(22,12)})
```

### Does Bra Size and Marriage Related?

```python
sns.set_style('whitegrid')
sns.countplot(x='Size',hue='Married?',data=train,palette='rainbow')
sns.set(rc={'figure.figsize':(22,12)})
```
